### 전체 데이터 중 방송명 분석 

```python
from konlpy.tag import Mecab
import re

# bn = 방송이름 
all_bn = All_data['방송명']

all_bn_clean = [ ]

#정규식으로 한글만 남기기 
for bn in all_bn:
  clean = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", " ", bn)
  all_bn_clean.append(clean)



from konlpy.tag import Mecab

tagger = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")

# 형태소 분석, 불용어 처리 
nouns_all_bn = [ ]
for bn in all_bn_clean:
  tagged = tagger.pos(bn)
  nouns = [s for s, t in tagged if t in ['NNG', 'NNP', 'VA', 'XR'] and len(s) >1]
  for noun in nouns:
    if noun not in stop_words:
        nouns_all_bn.append(noun)


```

```python
# 본문 빈도수 측정 
from collections import Counter
nouns_counter = Counter(nouns_all_bn)
top_bn = dict(nouns_counter.most_common(100))
```

```python
def make_wordcloud(df):
    from konlpy.tag import Mecab
    import re

    # bn = 방송이름 
    tmp = df['방송명']

    tmp_clean = [ ]

    #정규식으로 한글만 남기기 
    for bn in tmp:
        clean = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", " ", bn)
        tmp_clean.append(clean)


    from konlpy.tag import Mecab

    tagger = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")

    # 형태소 분석, 불용어 처리 

    tmp_bn = [ ]
    for bn in tmp_clean:
        tagged = tagger.pos(bn)
        nouns = [s for s, t in tagged if t in ['NNG', 'NNP', 'VA', 'XR'] and len(s) >1]
        for noun in nouns:
            if noun not in stop_words:
                tmp_bn.append(noun)

    
    # 본문 빈도수 측정 
    from collections import Counter
    tmp_counter = Counter(tmp_bn)
    top_tmp = dict(tmp_counter.most_common(100))


    wc_tmp = WordCloud(background_color = 'white', font_path = fontpath)
    wc_tmp.generate_from_frequencies(top_tmp) 

    figure =  plt.figure(figsize = (30, 30))
    ax =  figure.add_subplot(1, 1, 1)
    ax.axis('off')
    ax.imshow(wc_tmp)
    
    plt.show()

    
```

#### 키워드 분석 함수 

```python
# 키워드 분석 필요 

from gensim.models.ldamodel import LdaModel
from gensim.models.callbacks import CoherenceMetric
from gensim import corpora
from gensim.models.callbacks import PerplexityMetric

import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

import pickle
import pyLDAvis.gensim_models as gensimvis
import pyLDAvis
from gensim.models.coherencemodel import CoherenceModel
import matplotlib.pyplot as plt
```

```python
def keywords(df, topics):
  tmp = df.drop_duplicates('방송명')
  tmp = tmp['방송명']

  tmp_clean = [ ]
  #정규식으로 한글만 남기기 
  for bn in tmp:
    clean = re.sub("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", " ", bn)
    tmp_clean.append(clean)

  tagger = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")

  # 본문내용
  tmp_bn_lda = []
  
  for bn in tmp_clean:
    tagged = tagger.pos(bn)
    nouns = [s for s, t in tagged if t in ['NNG', 'NNP', 'VA', 'XR'] and len(s) >1]
    tmp_list = []
    for noun in nouns:
      if noun not in stop_words:
        tmp_list.append(noun)
        tmp_bn_lda.append(tmp_list)


  dictionary_tmp = corpora.Dictionary(tmp_bn_lda)

  # 50%이상 나오는 단어, 5번 이하 나오는 단어는 필터링 

  dictionary_tmp.filter_extremes(no_below=5, no_above=0.5)

  # 사전 속의 단어가 문장에서 몇 번 출현하는지 빈도를 세서 벡터화 합니다.
  corpus_tmp = [dictionary_tmp.doc2bow(text) for text in tmp_bn_lda]

  num_topics = topics
  passes = 20
  iterations = 400

  ldamodel_tmp = LdaModel(corpus_tmp,
                      num_topics = num_topics,
                      id2word=dictionary_tmp, 
                      iterations= iterations,
                      alpha='auto',
                      eta='auto',
                      eval_every=None,
                      passes=passes)

  top_topics = ldamodel_tmp.top_topics(corpus_tmp) #, num_words=20)

  # Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
  avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics




  lda_visualization = gensimvis.prepare(ldamodel_tmp, corpus_tmp, dictionary_tmp, sort_topics=False)
  pyLDAvis.save_html(lda_visualization, '{0}_{1}.html'.format(df, topics))
```
